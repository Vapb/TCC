```python
import pandas as pd
```

```python
df = pd.read_csv('poc_ALL.csv')
```

```python
print("INFO Dataset: ", df.shape)
df = df.drop(['numero'], axis = 1)
df.head()
```

```python
# Criar Data Frames do Benigno e Maligno
maligno = df.loc[df['Malware'] == 1]
benign = df.loc[df['Malware'] == 0]
# Drop malware collums
maligno = maligno.drop(['Malware'], axis = 1)
benign = benign.drop(['Malware'], axis = 1)
```

```python
maligno.head()
```

```python
benign.head()
```

```python
# DADOS SOBRE A BASE

size_t = df.shape[0]
size_m = maligno.shape[0]
size_b = benign.shape[0]

perm_total_t = (df.sum()).sum()
perm_total_m = (maligno.sum()).sum()
perm_total_b = (benign.sum()).sum()

perm_por_t = (perm_total_t)/(size_t)
perm_por_m = (perm_total_m)/(size_m)
perm_por_b = (perm_total_b)/(size_b)

print('A QUANTIDADE TOTAL DE APLICATIVO Ã‰ -> ' + str(size_t) + '.\n - O NUMERO DE PERMISSOES UTILIZADAS -> ' + str(perm_total_t) + '.\n - O NUMERO DE PERM POR APLICATIVO Ã‰ -> ' + str(perm_por_t) + '\n')
print('A QUANTIDADE TOTAL DE MALWARE Ã‰ -> ' + str(size_m) + '.\n - O NUMERO DE PERMISSOES UTILIZADAS -> ' + str(perm_total_m) + '.\n - O NUMERO DE PERM POR MALWARE Ã‰ -> ' + str(perm_por_m) + '\n')
print('A QUANTIDADE TOTAL DE BENIGNO Ã‰ -> ' + str(size_b) + '.\n - O NUMERO DE PERMISSOES UTILIZADAS -> ' + str(perm_total_b) + '.\n - O NUMERO DE PERM POR BENIGNO Ã‰ -> ' + str(perm_por_b) + '\n')
```
