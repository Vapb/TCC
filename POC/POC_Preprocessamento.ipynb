```python
import pandas as pd
```

```python
df = pd.read_csv('poc_ALL.csv')
```

```python
df.head()
```

```python
df = df.drop(['numero'], axis = 1)
```

```python
x = df.sum()
```

```python
l1 = x.tolist()
```

```python
l2 = df.columns.tolist()
```

```python
fd = pd.DataFrame(list(zip(l2, l1)),columns =['Name', 'val']) 
```

```python
fd
```

```python
maior  = fd.loc[fd['val'] > 1]
print("INFO Dataset: ", maior.shape)
```

```python
android = fd[(fd['Name'].str.contains("ANDROID"))]
print("INFO Dataset: ", android.shape)
```

```python
new = pd.concat([maior, android])
```

```python
new = new.drop_duplicates()
print("INFO Dataset: ", new.shape)
```

```python
# LISTA DE PERMISSOES PARA O POC MELHORADO (ANDROID E >2)
important = new['Name'].tolist()
```

```python
jumbo  = df[important]
print("INFO Dataset: ", jumbo.shape)
```

```python
# Remover possiveis linhas com apenas zeros
jumbo = jumbo[(jumbo.T != 0).any()]
```

```python
# Remover NULOS
jumbo = jumbo.dropna()
print("INFO Dataset: ", jumbo.shape)
```

```python
jumbo.head()
```

```python
jumbo.to_csv('poc_preprocessado.csv')
```
