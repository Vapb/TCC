# Teste para Tratamento de Dados
**Obj**: *Transformar uma lista dentro de uma coluna em colunas bin√°rias*

```python
import pandas as pd
import ast
```

```python
data = pd.read_csv("dados.csv", sep = ';')
```

```python
data.head()
```

```python
y = ((data['Cities']).tolist())
x = [] 
for i in y:
    x.append(ast.literal_eval(i))

df_cities = pd.DataFrame(x)
```

```python
df_cities
```

```python
cities_obj = df_cities.stack()
```

```python
cities_obj
```

```python
cities_df = pd.get_dummies(cities_obj)
```

```python
cities_df
```

```python
cities_df = cities_df.sum(level=0)
```

```python
cities_df
```

```python
cities_df = pd.concat([data, cities_df], axis=1)
```
