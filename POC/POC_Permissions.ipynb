```python
import pandas as pd
import ast
```

```python
data = pd.read_csv('POC_ALL_UNTREATED.csv', sep = ',')
```

```python
data.head()
```

```python
y = ((data['PERMISSIONS']).tolist())
x = [] 
for i in y:
    x.append(ast.literal_eval(i))

df_permissions = pd.DataFrame(x)
```

```python
df_permissions.head()
```

```python
permissions_obj = df_permissions.stack()
```

```python
permissions_obj.head(100)
```

```python
permissions_df = pd.get_dummies(permissions_obj)
permissions_df.head(150)
```

```python
permissions_df = permissions_df.sum(level=0)
permissions_df.head()
```

```python
permissions_df = pd.concat([data, permissions_df], axis=1)
```

```python
final = permissions_df.drop(['Permissions','PERMISSIONS','Numero','Name'],axis = 1)
```

```python
final
```

```python
final.to_csv('poc_ALL.csv')
```
